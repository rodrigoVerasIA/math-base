$$
ŷ = w1x1 + w2x2 + ... + wnxn + b
$$

$$
ŷ = W^TX + b
$$

$$
L(y, ŷ) = \frac{\sum_{n=1}(yi - ŷ)}{n}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# y = 2x + 1
# bias = 1
# Find w and b that minimize prediction error
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([3, 5, 7, 9, 11])
```

$$
θ = (X^TX)^-1X^Ty
$$

```python
# Close-form solution for linear regression
# add bias column (x0 = 1)
X_b = np.c_[np.ones((X.shape[0], 1)), X]  # add x0 = 1 to each instance
print("X_b:\n", X_b)

# Calculate the weights
theta_best = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
print("Theta best (weights):\n", theta_best)
```

```python
# Gradient descent solution
theta = np.random.randn(2, 1)  # random initialization of weights
print("Initial theta:\n", theta)
learning_rate = 0.01
epochs = 1000

# Convert y to column vector
y = y.reshape(-1, 1)
print(y)

for epoch in range(epochs):
  gradients = 2 / len(X_b) * X_b.T @ (X_b @ theta - y)
  theta -= learning_rate * gradients

print("Weights from GD:", theta)
```

```python
plt.scatter(X, y, color='blue')
plt.plot(X, X_b @ theta, color='red')
plt.title('Linear Regression Fit')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
